```python
import numpy as np
from sklearn.model_selection import train_test_split
from dl85 import DL85Classifier
from sklearn.metrics import confusion_matrix, precision_score, f1_score
import os
import time
```

```python
source_folder = "data/pydl8.5_datasets"
dest_folder = "data/pydl8.5_datasets_split"
log_filename = "dl85_py_result.txt"
```

```python
data_files = os.listdir(source_folder)
with open(log_filename, 'a') as logfile:
    logfile.write("name, dl_accu, dl_f1, dl_time_elapsed, dl_accu_train \n")
for this_file in data_files:  
    filename_stem = this_file.split('.')[0]
    X_train = np.genfromtxt(dest_folder + "/" + filename_stem + "_X_train.txt", delimiter = " ")
    y_train = np.genfromtxt(dest_folder + "/" + filename_stem + "_y_train.txt", delimiter = " ")
    X_test = np.genfromtxt(dest_folder + "/" + filename_stem + "_X_test.txt", delimiter = " ")
    y_test = np.genfromtxt(dest_folder + "/" + filename_stem + "_y_test.txt", delimiter = " ")
    start_time = time.time()
    clf = DL85Classifier(max_depth = 3, min_sup=5, time_limit=360)
    clf.fit(X_train, y_train)
    end_time = time.time()
    time_elapsed = end_time - start_time
    y_pred = clf.predict(X_test)
    confmat = confusion_matrix(y_test, y_pred)
    TN = confmat[0][0]
    TP = confmat[1][1]
    FP = confmat[0][1]
    FN = confmat[1][0]
    f1 = TP/(TP+(FN+FP)/2)
    accu = (TP + TN)/(TP+TN+FP+FN)
    # get training accuracy
    y_pred = clf.predict(X_train)  # test on training
    confmat = confusion_matrix(y_train, y_pred)  # test on training
    TN = confmat[0][0]
    TP = confmat[1][1]
    FP = confmat[0][1]
    FN = confmat[1][0]
    f1_train = TP/(TP+(FN+FP)/2)
    accu_train = (TP + TN)/(TP+TN+FP+FN)
    with open(log_filename, 'a') as logfile:
        logfile.write(filename_stem + ', ' + "{:.5f}, {:.5f}, {:.2f}, {:.5f}".format(accu, f1, time_elapsed, accu_train) + '\n')
```
